```python
import numpy as np, pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.infoclimat.fr/bulletins-speciaux-phenomenes-violents.html"
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
lines = soup.find("tr")
end = lines.contents[2]
for im in lines.td.children:
    val=(im.get('src')[-5])
    if val!='/':
        val = int(val)
        print(val)
for i,el in enumerate(lines.td.next_siblings):
    txt = el.get_text().split(" ")
    print(i,txt[2], txt[3], txt[4][:-1], txt[5])#, txt[5])
```

```python
mois = {"janvier":1, "février":2, "mars":3, "avril":4, "mai":5, "juin":6, "juillet":7, "août":8, "septembre":9, "octobre":10, "novembre":11, "décembre":12}
meteo = {"inondation":1, "orage":2, "tempête":3, "pluie": 4, "froid":5, "chaleur":6, "neige":7}
meteo["neige"]
```

```python
import datetime
lines = soup.find_all("tr")
meteo_event = pd.DataFrame(data= np.zeros((315, 10)), columns=["starting_date", "duration", "snow", "floods", "storm","hurricane", "rain", "cold", "hot", "comment"])
meteo_event.head()
```

```python
for k,line in enumerate(lines):
    dates=[]
    commentaire=[]
    labels= np.zeros(7)
    for i,el in enumerate(line.td.next_siblings):
        if i < 2:
            txt = el.get_text().split(" ")
            day = txt[2]
            if day =='1er':
                day=1
            d= datetime.datetime(int(txt[4][:-1]), mois[txt[3]], int(day))
            #date = txt[4][:-1]+'-'+str(mois[txt[3]])+'-'+txt[2]+' '+txt[5]
            dates.append(d)
        else :
            commentaire=(el.get_text())
    for im in line.td.children:
        val=(im.get('src')[-5])
        if val!='/':
            val = int(val)%7
            labels[val]=1
    meteo_event.iloc[k,0] = datetime.date.isoformat(dates[0])
    meteo_event.iloc[k,1] = (dates[1]-dates[0]).days
    meteo_event.iloc[k,2:9] = labels
    meteo_event.iloc[k,9] = commentaire

```

```python
def get_cat(spec):
    idx = []
    if "pluvieux" in spec : 
        idx.append(4)
    if "inondation" in spec : 
        idx.append(1)
    if "vent" in spec : 
        idx.append(3)
    if "froid" in spec : 
        idx.append(5)
    if 'neigeux' in spec : 
        idx.append(0)
    if "douceur" in spec or "chaleur" in spec : 
        idx.append(6)
    if "orage" in spec : 
        idx.append(2)
        
    labels=np.zeros(7)
    labels[idx]=1
    
    return(labels)
```
